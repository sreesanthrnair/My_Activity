```python
import numpy as np
import pandas as pd
```

```python
data['time'] = pd.to_datetime(data['time'])
```

```python
data.info()
data.isnull().sum()
```

```python
data = data.dropna(subset=['latitude', 'longitude', 'mag'])
data
data.info()
```

```python
data[data['mag']>=4]
```

```python
data['day_name'] = data['time'].dt.day_name()
data.head()
```

```python
data['severity_level']=np.where(data['mag'] < 4, 'Low',
                     np.where(data['mag'] > 6.0, 'High', 'Moderate'))
data
```

```python
data['place'].value_counts()
```

```python
earthquake_summary = data.groupby(data['time'].dt.date).agg({'depth':'max', 'mag':'mean'}).reset_index()
earthquake_summary.columns = ['date', 'max_depth', 'average_magnitude']
earthquake_summary
```

```python
data.to_csv('cleaned_earthquakes.csv')
data
```

```python
earthquake_summary.to_csv('earthquake_summary.csv')
earthquake_summary
```

```python
import sqlite3
conn=sqlite3.connect("earthquake_summary.db")
```

```python
data=pd.read_csv("/content/cleaned_earthquakes.csv")
data.to_sql("Earthquake_data",conn,index=False,if_exists="replace")
```
