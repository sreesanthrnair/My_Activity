```python
plt.scatter(superstore['Sales'], superstore['Profit'], color='purple')
plt.title('Sales vs. Profit', fontweight = 'bold')
plt.xlabel('Sales', fontweight= 'bold')
plt.ylabel('Profit', fontweight='bold')
```

```python
plt.bar(superstore['Region'], superstore['Sales'], color='purple')
plt.title('Sales by Region', size=12, fontweight='bold')
plt.xlabel('Region', fontweight='bold')
plt.ylabel('Sales', fontweight='bold')
plt.show()
```

```python
print('''The South region has the highest sales, followed by West, then Central, and finally East.
This means most of the business revenue comes from the southern part of the country, while the eastern region contributes the least.
''')
```

```python
# Analysis of monthly sales and profits by months
```

```python
superstore['Months'] = pd.to_datetime(superstore['Order Date']).dt.to_period('M')

monthly_sales = superstore.groupby('Months')[['Sales']].sum().reset_index()
monthly_profit = superstore.groupby('Months')[['Profit']].sum().reset_index()
monthly_sales['Months'] = monthly_sales['Months'].dt.to_timestamp()
monthly_profit['Months'] = monthly_profit['Months'].dt.to_timestamp()

fig, ax = plt.subplots(1, 2, figsize=(16, 8), sharex=True)

fig.suptitle('Monthly sales and profits', size=20, fontweight='bold')

ax[0].plot(monthly_sales['Months'], monthly_sales['Sales'],'o-', c='purple',)
ax[0].set_xlabel('Months', size=12, fontweight='bold')
ax[0].set_ylabel('Sales', size=12, fontweight='bold')
ax[0].grid(True)

ax[1].plot(monthly_profit['Months'], monthly_profit['Profit'], 'o-', c='brown')
ax[1].set_xlabel('Months', size=12, fontweight='bold')
ax[1].set_ylabel('Profit', size=12, fontweight='bold')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
print('''Over the years, both sales and profits show an overall upward trend, though profits are more variable with noticeable ups and downs.
Sales grow steadily, while profit spikes and dips suggest that not all high sales months are equally profitable.
This could be due to discounts, product mix, or operational costs.
''')
```

```python
# Sales vs profit by category
```

```python
sns.scatterplot(data=superstore, x='Sales', y='Profit', hue='Category', palette='BuPu')
plt.title('Sales vs Profit by Category')
plt.show()
```

```python
print('''Different product categories show different patterns.
Technology tends to have higher sales and profit, while Furniture shows more losses and scattered results.
Office Supplies sits somewhere in between.
This means not all categories perform equally—some bring in more profit even with similar sales.''')
```

```python
# Heatmap of profit by region and category
```

```python
pivot = pd.pivot_table(data=superstore, index='Region', columns='Category', values='Profit', aggfunc='sum')
sns.heatmap(pivot, fmt='.1f', annot=True, cmap='rocket')
plt.title('Profit by Region and Category', fontweight='bold')
plt.show()
```

```python
print('''- Technology is the most profitable category in every region, especially in the West and East.
Office Supplies also perform well, with strong profits in the East and Central.
Furniture shows mixed results—losses in Central, but decent profits elsewhere.
This means the business earns the most from Technology, while Furniture might need attention in certain regions.''')
```
