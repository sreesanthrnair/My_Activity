```python
#1) Import pandas, numpy

import pandas as pd
import numpy as np

```

```python
# 2) Load dataset using pd.read_csv()

url=('https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/IMDB-Movie-Data.csv')
movies=pd.read_csv(url)
movies.head()
```

```python
# 3) Display: shape, head, column names, datatypes, missing value

movies.shape
```

```python
movies.head()
```

```python
movies.columns
```

```python
# To find datatypes
movies.dtypes
```

```python
movies.isna().sum()
```

```python
# 4) Check and remove duplicates
# inplace=True is used for change original data set
movies.drop_duplicates(inplace=True)
movies
```

```python
# 5) Rename columns ("Title" > "Movie_Title")

movies.rename(columns={'Title':'Movie_Title'},inplace=True)
movies
```

```python
# 6) Get unique genres

unique=set()#used for store unique value
movies['Genre'].apply(lambda x: unique.update(x.split(',')))
unique
```

```python
# 7) Extract number of genres per movie

movies['Genre_count']=movies['Genre'].apply(lambda x: len(x.split(',')))
movies.head()
```

```python
# 8) Handle missing values in 'Revenue (Millions)' and 'Metascore

movies[['Revenue (Millions)','Metascore']].isnull().sum()
```

```python
# 9) Top 5 movies by Rating

top_5_movies = movies.sort_values('Rating', ascending=False).head(5)
top_5_movies
```

```python
# 10) Movies with highest/lowest revenue

revenue=movies.loc[movies['Revenue (Millions)'].idxmax()]
revenue
```

```python
revenue=movies.loc[movies['Revenue (Millions)'].idxmin()]
revenue
```

```python
# 11) Avg. rating by Director

director=movies.groupby('Director')['Rating'].mean()
director
```

```python
# 12) Movies per year

movie_of_year=movies['Year'].value_counts().sort_index()
movie_of_year
```

```python
# Step 13: Movies with Rating > 8 & Revenue > 100M

high_profit=movies[(movies['Rating'] > 8) & (movies['Revenue (Millions)'] > 100)]
high_profit
```

```python
# 14) Group by Director: avg Rating, Revenue

movies.groupby('Director')[['Rating','Revenue (Millions)']].mean().reset_index()
```

```python
# Step 15: Group by Year – Count & Average Rating

yearly_stats = movies.groupby('Year').agg(
    Movie_Count=('Movie_Title', 'count'),
    Avg_Rating=('Rating', 'mean')
).reset_index()
yearly_stats
```

```python
# 16 a: Create Profitability Column

movies['Profitability'] = movies['Revenue (Millions)'] / movies['Runtime (Minutes)']
movies
```

```python
# 16 b: Top 10 Most Profitable Movies

top_profitable = movies.sort_values(by='Profitability', ascending=False).head(10)
top_profitable
```

```python
# 17: Categorize Movies by Rating

def categorize_rating(rating):
    if rating < 6:
        return 'Poor'
    elif rating < 7.5:
        return 'Average'
    else:
        return 'Excellent'

movies['Rating_Category'] = movies['Rating'].apply(categorize_rating)
movies.head()
```

```python
# 18: Custom Scoring Function

def score_movie(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

movies['Score'] = movies.apply(score_movie, axis=1)
movies.head()
```
